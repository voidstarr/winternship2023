```python
import pandas as pd

# retrieved from https://data.noaa.gov/datasetsearch/
data = pd.read_csv("./day-2/data.csv", header=1)
data.head()
```

```python
from datetime import date, timedelta

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_raw = data.iloc[0].Date.split("-")
end_raw = data.iloc[-1].Date.split("-")

start_date = date(int(start_raw[0]), int(start_raw[1]), int(start_raw[2]))
end_date = date(int(end_raw[0]), int(end_raw[1]), int(end_raw[2]))
holes = 0
for single_date in daterange(start_date, end_date):
    fmt_date = single_date.strftime("%Y-%m-%d")
    if data[data.Date == fmt_date].empty:
        print(f"found hole at {fmt_date}")
        holes += 1

print(f"{holes} found")
```
